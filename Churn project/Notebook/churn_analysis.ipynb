## 📊 Task 3: Exploratory Data Analysis (EDA)
In this step, we analyze the dataset to identify patterns, trends, and relationships between variables.

### 📈 Distribution of Numerical Features
We use histograms to understand how numerical data is distributed.

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,8))
plt.show()
```

### 📊 Insights from Histograms

- Most customers are either new (low tenure) or long-term users
- Monthly charges are mostly in the mid to high range
- Total charges are skewed towards lower values, indicating many new customers
- Majority of customers are non-senior citizens

### 🔍 Churn Distribution
Analyzing how many customers have churned vs retained.

```python
df['churn'].value_counts()
```

### 📊 Churn Analysis Insights

- Total customers: ~7000
- Customers retained: 5163 (~73%)
- Customers churned: 1869 (~27%)

A significant portion of customers have churned, indicating potential issues in customer satisfaction or service quality. This suggests the need for strategies to improve customer retention.

### 📈 Churn Distribution Visualization
Visualizing churn vs non-churn customers.

```python
df['churn'].value_counts().plot(kind='bar')
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

### 💰 Churn vs Monthly Charges
Analyzing how monthly charges affect customer churn.

```python
import matplotlib.pyplot as plt

df.boxplot(column='monthlycharges', by='churn')
plt.title("Churn vs Monthly Charges")
plt.suptitle("")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")
plt.show()
```

### 📊 Insight: Monthly Charges vs Churn

The boxplot shows that customers with higher monthly charges tend to churn more compared to those with lower charges.

This suggests that pricing plays a significant role in customer retention, and customers may leave due to high costs.

Customers paying above average monthly charges are more likely to discontinue the service.

### 📄 Churn vs Contract Type
Analyzing how different contract types affect cṁustomer churn.

```python
import matplotlib.pyplot as plt

df.groupby('contract')['churn'].value_counts().unstack().plot(kind='bar')
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Count")
plt.show()
```

### 📊 Insight: Contract Type vs Churn

Customers with month-to-month contracts have the highest churn rate, indicating that lack of long-term commitment leads to higher customer loss.

In contrast, customers with one-year and two-year contracts show significantly lower churn, suggesting that long-term contracts improve customer retention.

Long-term contracts act as a strong retention strategy for telecom companies.

### 💳 Churn vs Payment Method
Analyzing how different payment methods influence customer churn.

```python
df.groupby('paymentmethod')['churn'].value_counts().unstack().plot(kind='bar')
plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

### 📊 Insight: Payment Method vs Churn

Customers using electronic check have the highest churn rate, indicating that this payment method may lead to dissatisfaction or inconvenience.

On the other hand, customers using automatic payment methods (bank transfer or credit card) show lower churn, suggesting better customer retention.

### 👥 Customer Segmentation by Tenure
Grouping customers based on how long they have stayed with the company.

```python
df.loc[:, 'tenure_group'] = pd.cut(df['tenure'],
                                  bins=[0,12,36,72],
                                  labels=['0-12 Months','13-36 Months','37+ Months'])

df['tenure_group'].value_counts()
```

### 📊 Insight: Customer Segmentation

- Most customers belong to the long-term category (37+ months)
- A significant number of new customers (0–12 months) are also present
- Mid-term customers (13–36 months) are comparatively fewer

### 🥧 Customer Distribution by Tenure
Visualizing customer segmentation using a pie chart.

```python
df['tenure_group'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Segmentation by Tenure")
plt.ylabel("")
plt.show()
```

### 📊 Average Monthly Charges by Tenure Group
Analyzing how monthly charges vary across different customer segments.

```python
df.groupby('tenure_group', observed=False)['monthlycharges'].mean().plot(kind='bar')
plt.title("Average Monthly Charges by Tenure Group")
plt.xlabel("Tenure Group")
plt.ylabel("Average Monthly Charges")
plt.show()
```

### 📊 Insight: Charges vs Tenure

Customers with longer tenure tend to have higher average monthly charges. This indicates that loyal customers are more likely to subscribe to premium or higher-value services.

This highlights the importance of retaining long-term customers for increased revenue.
